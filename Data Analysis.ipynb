# 라벨링 데이터❗
#### 그 중 D08-1-033.json 파일 내용을 불러오는 방법입니다.

```python
path = './data/labeling_data/'  # 라벨링 데이터의 경로 지정
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.json')] # 해당 경로의 json 파일들을 목록화
```

```python
file_list_py # './data/labeling_data/' 디렉토리 가보면 이러한 json 파일들이 있을 겁니다.
```

```python
print(path+file_list_py[0]) # 그 중 첫번째 파일에 대한 예시입니다.
data1 = json.load(open(path+file_list_py[0], 'rt', encoding='UTF8')) # 인코딩 형식을 저런식으로 바꿔주지 않으면 오류가 납니다.
df = pd.DataFrame(data1["frames"]) # 제일 크게 frames, type, type_info에 대한 데이터들이 있는데 그 중 frames 인덱스에 대한 내용을 데이터프레임으로 나타내겠습니다.
df
```
